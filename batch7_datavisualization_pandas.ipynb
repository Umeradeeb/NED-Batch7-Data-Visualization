```python
# prompt: create 5 columns dataframe columns "id","student name", "father's Name","course" and "fee", 20 samples, names type should be pakistani

import pandas as pd
import random

# List of Pakistani names
names = [
    "Ali", "Ahmed", "Khan", "Hussain", "Malik", "Raza", "Hassan", "Zafar", "Asif", "Imran",
    "Fatima", "Aisha", "Sana", "Maria", "Saima", "Zara", "Rida", "Amina", "Saba", "Hina"
]

# Create a list to store the data
data = []

# Generate 20 samples
for i in range(20):
  student_name = random.choice(names) + " " + random.choice(names)
  father_name = random.choice(names) + " " + random.choice(names)
  course = random.choice(["Data Science", "AI", "Machine Learning", "Web Development"])
  fee = random.randint(1000, 5000)
  data.append([i + 1, student_name, father_name, course, fee])

# Create the DataFrame
df = pd.DataFrame(data, columns=["id", "student name", "father's Name", "course", "fee"])

# Print the DataFrame
df
```
