```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')
display(titanic.head())

# Set the size of the plot
plt.figure(figsize=(10, 6))

# Create a heatmap to visualize missing data
sns.heatmap(titanic.isnull(), cmap='viridis', cbar=False, yticklabels=False)

# Set titles and labels
plt.title('Missing Data in Titanic Dataset')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()

```

# Question: Which columns have the most missing data?

Deck has the most missing data, followed by Age and then Embarked.




```python
# Calculate the number of missing values in each column
missing_data = titanic.isnull().sum()

# Sort the columns by the number of missing values
missing_data_sorted = missing_data[missing_data > 0].sort_values(ascending=False)

# Display columns with the most missing data
print(missing_data_sorted)
```
