### Slicing
[start : end : step] for every dimension

* start = included
* end = excluded n-1
* step = sequence (default 1)

```python
#---> left to right start with zero
#             0. 1. 2. 3. 4  5
a = np.array([1, 2, 3, 4, 5, 6])
#            -6 -5.-4  -3 -2 -1      <------- right to left start -1,-2...
display(a)
display(a[-1])
display(a[0:2])
display(a[2:5])
display(a[:3])    # if there is no start value, then it consider all values 
display(a[:])
display(a[::])
```

```python
# slicing alway move from left to right if you pass positive or negative indexes
#---> left to right start with zero
#             0. 1. 2. 3. 4  5
a = np.array([1, 2, 3, 4, 5, 6])
#            -6 -5.-4  -3 -2 -1      <------- right to left start -1,-2...
display(a)
display(a[-5:4])
```

```python
# slicing alway move from left to right if you pass positive or negative indexes
#---> left to right start with zero
#             0. 1. 2. 3. 4  5
a = np.array([1, 2, 3, 4, 5, 6])
#            -6 -5.-4  -3 -2 -1      <------- right to left start -1,-2...
display(a)
display(a[::1])
display(a[::2])
display(a[::3])

display(a[::-1])
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load the image using PIL
image = Image.open('my_image.jpeg')

# Convert the image to a NumPy array
image_np = np.array(image)

display(image_np.shape)

# # Define the crop dimensions
# # For example, crop the image to a rectangle starting at (100, 100)
# # and going to (400, 400)
cropped_image_np = image_np[100:400, 100:400]

display(cropped_image_np.shape)


# Alternatively, display using matplotlib
plt.imshow(cropped_image_np)
plt.axis('off')  # Hide axis
plt.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from scipy.ndimage import convolve

# Define the box filter
filter = np.array([
    [1, 1, 1],
    [1, 1, 1],
    [1, 1, 1]
], dtype=np.float32) / 9.0  # Normalize the filter

# Load the image using PIL
image = Image.open('2pic.jpg')
image_arr = np.array(image) / 255.0  # Normalize the image array to [0, 1] range

# Apply the convolution filter to each channel separately
convolved_arr = np.zeros_like(image_arr)
for i in range(3):  # Assuming RGB channels
    convolved_arr[:, :, i] = convolve(image_arr[:, :, i], filter, mode='constant', cval=0.0)

# Convert the convolved array back to an image
convolved_image = Image.fromarray(np.uint8(255 * convolved_arr), 'RGB')  # Convert back to [0, 255] range

# Display the convolved image
plt.imshow(convolved_image)
plt.axis('off')  # Hide axis
plt.show()

plt.imshow(image)
plt.axis('off')  # Hide axis
plt.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from scipy.ndimage import convolve

# Define the box filter
filter =np.array([
    [1, 2, 1],
    [2, 4, 2],
    [1, 2, 1]
], dtype=np.float32) / 16.0  # Glow

# Load the image using PIL
image = Image.open('2pic.jpg')
image_arr = np.array(image) / 255.0  # Normalize the image array to [0, 1] range

# Apply the convolution filter to each channel separately
convolved_arr = np.zeros_like(image_arr)
for i in range(3):  # Assuming RGB channels
    convolved_arr[:, :, i] = convolve(image_arr[:, :, i], filter, mode='constant', cval=0.0)

# Convert the convolved array back to an image
convolved_image = Image.fromarray(np.uint8(255 * convolved_arr), 'RGB')  # Convert back to [0, 255] range

# Display the convolved image
plt.imshow(convolved_image)
plt.axis('off')  # Hide axis
plt.show()

plt.imshow(image)
plt.axis('off')  # Hide axis
plt.show()
```
